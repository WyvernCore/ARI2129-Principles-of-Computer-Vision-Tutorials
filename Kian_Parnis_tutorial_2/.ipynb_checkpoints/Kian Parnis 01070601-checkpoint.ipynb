# Morphology Tutorial

Mathematical morphology (MM) is a theoretical model for digital images built upon lattice theory and topology. It is the foundation of morphological image processing, which is based on shift-invariant (translation invariant) operators.Originally developed for binary images it has now been successfully extend to grey scale images.

```python
coin = cv2.imread('images/Euro_Coins.jpg')
grayCoin = cv2.cvtColor(coin, cv2.COLOR_BGR2GRAY)
```

printImg is a helper function used to display the images, and if needed the histograms aswell.

```python
def printImg(original, new, showHist):
    fig = plt.figure(figsize=(18, 4))

    fig.add_subplot(1, 2, 1)
    
    plt.title("First")
    plt.imshow(original)
    plt.axis('off')
    
    fig = plt.figure(figsize=(18, 4))
    fig.add_subplot(1, 2, 2)
    
    plt.title("Second")
    plt.imshow(new)
    plt.axis('off')
    if(showHist):
        colors = ('r','g','b')
        fig = plt.figure(figsize=(18, 4))
        fig.add_subplot(1, 2, 1) 

        for i,color in enumerate(colors):
            hist = cv2.calcHist([original],[i],None,[256],[0,256])
            plt.plot(hist,color = color) 
    
        plt.title("First")    
        plt.show()

        fig = plt.figure(figsize=(18, 4))
        fig.add_subplot(1, 2, 2)

        for i,color in enumerate(colors):
            histTwo = cv2.calcHist([new],[i],None,[256],[0,256])
            plt.plot(histTwo, color = color) 
        
        plt.title("Second")    
        plt.show()  
```

Connectivity was tested using the below function, this was tested with various means such as allpying a gaussianblur to lower the noise varying the lowerbounf of the threshhold values and testing with connectivity 4 and 8.

The best results appeared at the below configuration.

```python
def connectedComp(passedIm, ColourMin, orig, showHist):
    blurIm = cv2.GaussianBlur(passedIm, (11,11), cv2.BORDER_DEFAULT)
    ret, threshImage = cv2.threshold(blurIm,ColourMin,255, cv2.THRESH_BINARY_INV)
   
    ret2, labels = cv2.connectedComponents(threshImage, connectivity=8)

    labelH= np.uint8(172 * labels / np.max(labels))
    blank_ch = 220 * np.ones_like(labelH) #Control brightness of image
    labeledIm = cv2.merge([labelH, blank_ch, blank_ch]) #Merge images together
    labeledIm = cv2.cvtColor(labeledIm, cv2.COLOR_HSV2BGR)
    labeledIm[labelH == 0] = 0

    printImg(orig[:,:,::-1], labeledIm, showHist)
    
connectedComp(grayCoin, 225, coin, False)
```

<b style="font-size:18px;">Exercise 1.1 - Dilation</b>

Download the image file handwritten text.png from VLE. Apply the opencv method for dilation (dilate()). Vary the dilation filter and compare results using histograms. What can you observe?

```python
textIm = cv2.imread('images/text.png')
textIm = textIm[:,:,::-1]
```

Dilation was implemented using varying matrix sizes, and the results where compared using histograms.
The higher the dilation value the more the area around the texts starts to blend with the text.
At a very high dilation value eg: 70, the lower brightness values get pruned out, leading to ones between ~190 - 255 being kept.

```python
def dilate(img,matrixSize):
    kernel = np.ones((matrixSize, matrixSize), 'uint8')
    dilation = cv2.dilate(img,kernel, iterations=1)
    printImg(img, dilation, True)
    
dilate(textIm, 3)
```

<b style="font-size:18px;">Exercise 1.2 - Erosion</b>

Using the same text based image from the previous exercise apply the opencv method for erosion (erode()). Vary the erosion filter and compare results using histograms. What can you observe?

Erosion, as expected has the inverted effect dilate does and expands the image's text to the rest of the image area.
Due to the text image being at a brightness of 0, a very eroded image in this case would lead to the histograms values accumulating at position 0 of x.

```python
def Erosion(img,matrixSize):
    kernel = np.ones((matrixSize, matrixSize), 'uint8')
    erosion = cv2.erode(img,kernel)
    printImg(img, erosion, True)

Erosion(textIm, 7)
```

<b style="font-size:18px;">Exercise 1.3 - Opening</b>

Opening is an applied process of erosion followed by dilation. It is typically used to segment neck parts of shapes found within an image. 

It is also used in Text segmentation to seperate lines and words and sometimes characters.
Using the same text image from previous exercises apply this process of opening to visually separate lines words and characters.


```python
def opening(img,matrixSize):
    kernel = np.ones((matrixSize, matrixSize), 'uint8')
    opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    printImg(img, opening, False)
    
opening(textIm, 3)
```

<b style="font-size:18px;">Exercise 1.4 - Closing</b>

Closing is an applied process of dilation followed by erosion. It is typically used to fill in shapes found within an image. It is also used in Text segmentation to reinforce eroded characters.

Using the same text image from previous exercises apply this process of opening to visually separate lines words and characters.


```python
def closing(img,matrixSize):
    kernel = np.ones((matrixSize, matrixSize), 'uint8')
    closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    printImg(img, closing, False)
    
closing(textIm, 3)
```

Using histograms, apply any of the morphological techniques used in the previous exercises to try and segment the text into seperate lines. The lines do not have to be perfect. 

```python
def Dilate(img,matrixSize):
    kernel = np.ones((matrixSize, matrixSize), 'uint8')
    dilation = cv2.dilate(img,kernel)
    return dilation 
```

```python
def erosion(img,matrixSize):
    kernel = np.ones((matrixSize, matrixSize), 'uint8')
    erosion = cv2.erode(img,kernel)
    return erosion
```

With the use of histograms to better predict how the previous morphological techniques where used to attempt to seperate each line.
This wasnt fully successfull due to certain letters writtien in such a way, where performing opening / closing almost always lead to some sort of word being segmented to the word below it.
In the end, performing an opening with kernel=2 and connected components with ThreshHold min=165 seems to be the best attempt I could make at segmenting each line.

```python
textIm = cv2.imread('images/text.png')
textGray = cv2.cvtColor(textIm, cv2.COLOR_BGR2GRAY)

er = erosion(textGray, 2)
di = Dilate(er, 2)

connectedComp(di, 165, textIm, True)
```
